```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

## Import Data Set

```python
iris = datasets.load_iris()
X = iris.data[:, :2] # Set X varibles
y = iris.target # Set target variables

# Initial Visualization of the Iris Data Set
plt.scatter(X[:,0], X[:,1], c=y, cmap = 'gist_rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
# Goal of K-Means is to classify the Iris's into categories based on there sepal width and heigh
# Creating Kmeans Object using KMeans()
kmean = KMeans(n_clusters=3, random_state=1)

# Fit on data
kmean.fit(X)
KMeans(algorithm = 'auto', 
        copy_x=True,
        init='k-means++', # selects initial cluster centers
        max_iter=300,
        n_clusters=3,
        n_init=10,
        n_jobs=None,
        precompute_distances='auto',
        random_state=1,
        tol=.0001, # min. tolerance for distance between clusters
        verbose=0)

# Plotting cluster centers 
# initiate a variable for the centers
centers = kmean.cluster_centers_
print(centers)
```

## Implementation of the K-Means Centers on the Original Data Set
Based on the two figures it is obvious that K-Means picks up the Cluster in Red (Green), but has difficulty segmenting clusters in teal (brown) and pink (blue), this is because of the overlap inherent to those groupings. <br />
Algorithm behind KMeans Clustering <br />
1: Select some value for K (Number of Clusters) <br />
2: Initialize K "centroids" or the starting points in your data <br />
3: Create the clusters, assigning each point to the nearest cluster centroid <br />
4: Improve each cluster by moving each centroid to the center of its cluster <br />
5: Repeat steps 3 and 4 until the centroids converge (The assignments no longer change)

```python
new_labels = kmean.labels_
# Plot the identified clusters and compare
fig, axes = plt.subplots(1,2,figsize=(12,7))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow', 
                edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet', 
                edgecolor='k', s=150)
axes[0].set_xlabel('Sepal Length')
axes[1].set_xlabel('Sepal Length')
axes[0].set_ylabel('Sepal Width')
axes[1].set_ylabel('Sepal Width')
axes[0].set_title('Original')
axes[1].set_title('Predicted')
```
